# Functions in Figure-level vs Axes-level

- Seaborn has several functions in different levels:

    1. Figure-level
    2. Axes-level


- Axes-level functions will draw plots into one `matplotlib.pyplot.Axes`.

- Figure-level functions will interact with figures in Matplotlib through a Seaborn object.

- Every module has has one figure-level function which includes interfaces for axes-level functions.

    1. relplot (Relational):
        - scatterplot
        - lineplot
        
    2. displot (Distribution):
        - histplot
        - kdeplot
        - ecdfplot
        - rugplot
        
    3. catplot (Categorical):
        - stripplot
        - swarmplot
        - boxplot
        - violinplot
        - pointplot
        - barplot


Documentation: https://seaborn.pydata.org/tutorial/function_overview.html

# relplot vs scatterplot & lineplot

## relplot

- The `replot()` function gives access to some axes-level functions: `scatterplot()` and `lineplot()`.

- To draw a plot for each axes-level function, use parameter `kind=`.

`scatterplot()`: with `kind="scatter"`

`lineplot()` : with `kind="line"`
    

- Documentation: https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

```python
sns.relplot(data=tips,
           x='total_bill',
            y='tip',
            hue='sex',
            style='sex',
            size='size',
#             kind='scatter'
           )
```

# Displot vs Histplot

- To use distribution plots, use these parameters:

    1. `histplot()` (with `kind="hist"`)

    2. `kdeplot()` (with `kind="kde"`)

    3. `ecdfplot()` (with `kind="ecdf"`)


Documentation: https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot

```python
sns.distplot(data=tips,
            x="total_bill",
#             col="time",
#             kde=True,
#             hue="smoker",
#             rug=True,
            kind='ecdf'
           )
```

# Catplot vs Barplot

Categorical scatterplots:

- `stripplot()` (with `kind="strip"`;  default)

- `swarmplot()` (with `kind="swarm"`)

Categorical distribution plots:

- `boxplot()` (with `kind="box"`)

- `violinplot()` (with `kind="violin"`)

- `boxenplot()` (with `kind="boxen"`)

Categorical estimate plots:

- `pointplot()` (with `kind="point"`)

- `barplot()` (with `kind="bar"`)

- `countplot()` (with `kind="count"`)


Documentation: https://seaborn.pydata.org/generated/seaborn.catplot.html#seaborn.catplot

```python
sns.catplot(data=tips,
            kind="violin",
            x="day",
            y="total_bill",
            hue="smoker",
#             dodge=True,
           )
```

```python
sns.boxenplot(data=tips,
#             kind="boxen",
            x="day",
            y="total_bill",
            hue="smoker",
#               kind='violin'
             )
```

```python
# Quiz

# violinplot
sns.catplot(data=tips,
            kind="violin",
            x="day",
            y="total_bill",
            hue="smoker"
           )

sns.catplot(data=tips,
            kind="violin",
            x="day",
            y="total_bill",
            hue="smoker"
           )



# boxplot
sns.catplot(data=tips,
            kind="box",
            x="day",
            y="total_bill",
            hue="smoker"
           )

sns.catplot(data=tips,
            kind="box",
            x="day",
            y="total_bill",
            hue="smoker"
           )
```
