# Line and Multiline Plot

## Line plot

- A line plot can help to show frequency over a continuous variable; e.g.: frequency over time

- To make a line plot, use `plt.plot()` with variables that want to be drawn in x-axis and y-axis.


- Value `b-` is used to give the color blue. Try other colors as well!


- In the mockup data below, there is a positive cases of a disease from March to June. Draw a line plot to show frequency over time!

```python
# value list for x-axis
x = ['Maret', 'April', 'Mei', 'Juni']

# value list for y-axis
y = [5, 500, 750, 1500]

# Draw the line plot
plt.plot(x, y, 'b--')

# Show the plot
plt.show()
```

## Multiline Plot

- A multiline plot is used to show line plot with different category variables.


- The example below shows the mock-up data of positive cases of diseases from March to June for every age category.

```python
# list value for x-axis
x = ['Maret', 'April', 'Mei', 'Juni']

# list value for y-axis
y1 = [0, 10, 300, 400]  # group age 0-17 years old
y2 = [4,  340, 400, 600] # group age 18-59 years old
y3 = [1, 150, 50, 500] # group age >= 60 years old

# Draw the plot
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

# Show the plot
plt.show()
```

# Histogram

- A histogram us used to show the distribution of a numeric variable.

- Histogram uses binning with the same range. The height of the bar shows the frequency of the binned data group.

- To show histogram, use `plt.hist()`.

```python
# Random numeric data from 0 to 100 with 50 data points
x = np.random.randint(100, size=50)

print(x)

# Gambarkan histogram
plt.hist(x, bins=10)

# Tunjukkan plot yang telah digambar
plt.show()
```

# Bar Chart

- A bar chart is used in a categorical data where the height of each bar represent the frequency of each category.

- To draw a bar chart, use `plt.bar()`.

- One example of when to use a bar chart is to show the frequency of the sales of an online shop for each catalog.

```python
# value for catalog
katalog = ['makanan', 'pakaian', 'gadget']

# value for sales frequency
jual = [240, 120, 75]

# Draw the bar chart
plt.bar(katalog, jual)

# Show the plot
plt.show()
```

## Horizontal Bar Chart

- A bar chart does not always have to be vertical.

# Boxplot

- A boxplot is used to show a more contextual distribution through quantiles, minimum, and maximum values.

- To draw a boxplot, use `plt.boxplot()`.



image.png

- The example below shows a boxplot using 100 random data from 0 to 100.

```python
# value x
x = np.random.randint(100, size=100)

# show value x
print(x)

# Draw the boxplot
plt.boxplot(x)

# Show the plot
plt.show()
```

```python
B = plt.boxplot(x)
[item.get_ydata() for item in B['whiskers']]
```

```python
[item.get_ydata() for item in B['whiskers']]
```

```python
print(np.quantile(x, 0.25)) #quantile 1
print(np.quantile(x, 0.5)) #quantile 2
print(np.quantile(x, 0.75)) #quantile 3

print(min(x)) # min
print(max(x)) # max
```
