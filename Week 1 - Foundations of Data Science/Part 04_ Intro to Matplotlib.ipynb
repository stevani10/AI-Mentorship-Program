## 2. Matplotlib Object Hierarchy

- This is the hierarchy of an object in Matplotlib:

  1. Figure
  2. Axes

- A figure is the outermost holder of a plot. A figure may include several axes.

- Axes represent a plot that includes axes, tick marks, lines, legends, titles, and text-boxes.

## 3. Pyplot Interface

- There are several interfaces of Matplotlib:

    1. Pyplot interface
    2. Object-oriented interface
    3. Pylab interface (not discussed in this course)


- The first one and the one that is mostly used is pyplot interface.

- The pyplot interface looks like this:

```python
fig, ax = plt.subplots()
plt.show()
```
