# Intro to Seaborn

- Seaborn is a library to create statistical graph.

- Seaborn is integrated to Pandas data structure to make it easier to create graphs.

- Documentation: https://seaborn.pydata.org/introduction.html

# Seaborn and Matplotlib Differences

1. Use
    - Matplotlib: customizable plotting.
    - Seaborn: Has varieties of plots, focusing on statistical plots.
    
    
2. Visualization
    - Matplotlib: more customizable, many functions to manipulate every aspect of a plot.
    - Seaborn: More integrated with Pandas. Help to focus on the elements of the plots.

# Data Structure that can be used in Seaborn

- There are two types of datasets that can be processed by Seaborn:

    1. Long-form data
    2. Wide-form data

Documentation: https://seaborn.pydata.org/tutorial/data_structure.html

```python
# Example of Long-form data

flights = sns.load_dataset("flights")
flights.head(10)
```

```python
# Example of Wide-form data

flights_wide = flights.pivot(index="year", columns="month", values="passengers")
flights_wide.head()
```
